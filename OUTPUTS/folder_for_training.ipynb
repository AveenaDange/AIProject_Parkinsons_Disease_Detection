```python
import os
import shutil
import random

# Define paths for your augmented data
augmented_folder = r"C:\Users\ambik\Desktop\AI Project\Project Files\augmented"  # Adjust this path as necessary
base_output_folder = r"C:\Users\ambik\Desktop\AI Project\Project Files\data2"  # Desired base output path

# Define the folder structure
folder_structure = {
    'train': {
        'healthy': ['spiral', 'meander', 'circle'],  # Adjust based on your available folders
        'pd': ['spiral', 'meander', 'circle']
    },
    'validation': {
        'healthy': ['spiral', 'meander', 'circle'],
        'pd': ['spiral', 'meander', 'circle']
    }
}

# Create the output folder structure
for category, subfolders in folder_structure.items():
    for health_status, tasks in subfolders.items():
        for task in tasks:
            # Create directories
            path = os.path.join(base_output_folder, category, health_status, task)
            os.makedirs(path, exist_ok=True)

# Function to move files from augmented dataset to the new structure
def move_files(dataset_name, health_status):
    dataset_path = os.path.join(augmented_folder, dataset_name)
    
    # Iterate through the types of drawings
    for task in ['spiral', 'meander', 'circle']:
        source_folder = os.path.join(dataset_path, task, health_status)
        
        # Check if the source folder exists
        if os.path.exists(source_folder):
            print(f"Processing files from: {source_folder}")  # Debug statement
            
            # Get all image files
            image_files = [f for f in os.listdir(source_folder) if f.endswith('.jpg') or f.endswith('.png')]
            random.shuffle(image_files)  # Shuffle the list to ensure random distribution
            
            # Split into training and validation sets (80% train, 20% validation)
            split_index = int(len(image_files) * 0.8)  # 80% for training
            train_files = image_files[:split_index]
            val_files = image_files[split_index:]
            
            # Move training files
            for filename in train_files:
                shutil.move(os.path.join(source_folder, filename), os.path.join(base_output_folder, 'train', health_status, task, filename))
                
            # Move validation files
            for filename in val_files:
                shutil.move(os.path.join(source_folder, filename), os.path.join(base_output_folder, 'validation', health_status, task, filename))
            
            print(f"Moved {len(train_files)} files to train/{health_status}/{task}")
            print(f"Moved {len(val_files)} files to validation/{health_status}/{task}")

        else:
            print(f"Source folder does not exist: {source_folder}")  # Debug statement

# Move files from each dataset
move_files('HandPD', 'healthy')
move_files('HandPD', 'pd')
move_files('NewHandPD', 'healthy')
move_files('NewHandPD', 'pd')
move_files('ParkinsonsDrawing', 'healthy')
move_files('ParkinsonsDrawing', 'pd')

print("Folder structure created and files moved successfully!")
```

```python
import os
import shutil
import random

# Define paths
augmented_folder = r"C:\Users\ambik\Desktop\AI Project\Project Files\augmented"
destination_folder = r"C:\Users\ambik\Desktop\AI Project\Project Files\data2"

# Create train and validation folders if they don't exist
train_folder = os.path.join(destination_folder, 'train')
validation_folder = os.path.join(destination_folder, 'validation')

os.makedirs(train_folder, exist_ok=True)
os.makedirs(validation_folder, exist_ok=True)

# Function to move files from augmented to data2
def move_files():
    # Process HandPD
    for task in ['spiral', 'meander']:
        for health_status in ['healthy', 'pd']:
            source_folder = os.path.join(augmented_folder, 'HandPD', task, health_status)
            if os.path.exists(source_folder):
                files = os.listdir(source_folder)
                random.shuffle(files)
                split_index = int(0.8 * len(files))  # 80% for training

                # Move files to the training set
                for file in files[:split_index]:
                    move_file(source_folder, file, train_folder, health_status, task)

                # Move files to the validation set
                for file in files[split_index:]:
                    move_file(source_folder, file, validation_folder, health_status, task)

    # Process NewHandPD
    for task in ['spiral', 'meander', 'circle']:
        for health_status in ['healthy', 'pd']:
            source_folder = os.path.join(augmented_folder, 'NewHandPD', task, health_status)
            if os.path.exists(source_folder):
                files = os.listdir(source_folder)
                random.shuffle(files)
                split_index = int(0.8 * len(files))  # 80% for training

                # Move files to the training set
                for file in files[:split_index]:
                    move_file(source_folder, file, train_folder, health_status, task)

                # Move files to the validation set
                for file in files[split_index:]:
                    move_file(source_folder, file, validation_folder, health_status, task)

    # Process ParkinsonsDrawing
    for task in ['spiral', 'wave']:
        for health_status in ['healthy', 'pd']:
            source_folder = os.path.join(augmented_folder, 'ParkinsonsDrawing', 'training', task, health_status)
            if os.path.exists(source_folder):
                files = os.listdir(source_folder)
                random.shuffle(files)
                split_index = int(0.8 * len(files))  # 80% for training

                # Move files to the training set
                for file in files[:split_index]:
                    move_file(source_folder, file, train_folder, health_status, task)

                # Move files to the validation set
                for file in files[split_index:]:
                    move_file(source_folder, file, validation_folder, health_status, task)

# Function to move a file and handle name conflicts
def move_file(source_folder, file, dest_folder, health_status, task):
    src_path = os.path.join(source_folder, file)
    dest_path = os.path.join(dest_folder, health_status, task, file)

    # Create health status and task directories if they don't exist
    os.makedirs(os.path.join(dest_folder, health_status, task), exist_ok=True)

    # Check for conflicts and rename if necessary
    if os.path.exists(dest_path):
        base, extension = os.path.splitext(file)
        counter = 1
        new_file_name = f"{base}_{counter}{extension}"
        new_dest_path = os.path.join(dest_folder, health_status, task, new_file_name)

        # Increment until a unique name is found
        while os.path.exists(new_dest_path):
            counter += 1
            new_file_name = f"{base}_{counter}{extension}"
            new_dest_path = os.path.join(dest_folder, health_status, task, new_file_name)

        dest_path = new_dest_path  # Update dest_path to the new unique name

    shutil.move(src_path, dest_path)
    print(f'Moved {file} to {dest_path}')

# Call the function
move_files()
```
